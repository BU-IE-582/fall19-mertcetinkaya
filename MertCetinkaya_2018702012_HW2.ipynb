```r
#Let's read my photo, print its dimensions and print that photo.
library(jpeg)
photo=readJPEG("C:/Users/mertc/Desktop/boun/2019fall/data_mining/week5/hw2/data/mert_image.jpg")
dim(photo)
plot(0:256,0:256,type="n",axes=TRUE)
rasterImage(photo,0,0,256,256)
```

```r
#Let's produce our noisy photo and print it.
noisy_photo_red=photo[,,1]+runif(256*256,0.1*max(photo[,,1]),min(photo[,,1]))
noisy_photo_green=photo[,,2]+runif(256*256,0.1*max(photo[,,2]),min(photo[,,2]))
noisy_photo_blue=photo[,,3]+runif(256*256,0.1*max(photo[,,3]),min(photo[,,3]))
noisy_photo=photo
noisy_photo[,,1]=noisy_photo_red
noisy_photo[,,2]=noisy_photo_green
noisy_photo[,,3]=noisy_photo_blue
noisy_photo=ifelse(noisy_photo>1,1,noisy_photo)
plot(0:256,0:256,type="n",axes=TRUE)
rasterImage(noisy_photo,0,0,256,256)
```

```r
#let's produce and print each channel of our noisy photo.
par(mfrow = c(2,3))
#We print the results of image function. 
image(0:256,0:256,t(apply(noisy_photo_red,2,rev)),  axes = TRUE, main = "Red Channel" , xlab = "",ylab = "")
image(0:256,0:256,t(apply(noisy_photo_green,2,rev)),  axes = TRUE, main = "Green Channel" , xlab = "",ylab = "")
image(0:256,0:256,t(apply(noisy_photo_blue,2,rev)),   axes = TRUE, main = "Blue Channel", xlab = "",ylab = "" )
#We get each layer or channel and fill the others with 0 manually for each color and print the results.
noisy_photo[,,1]=noisy_photo_red
noisy_photo[,,2]=0
noisy_photo[,,3]=0
noisy_photo=ifelse(noisy_photo>1,1,noisy_photo)
plot(0:256,0:256,type="n",axes=TRUE)
rasterImage(noisy_photo,0,0,256,256)

noisy_photo[,,1]=0
noisy_photo[,,2]=noisy_photo_green
noisy_photo[,,3]=0
noisy_photo=ifelse(noisy_photo>1,1,noisy_photo)
plot(0:256,0:256,type="n",axes=TRUE)
rasterImage(noisy_photo,0,0,256,256)

noisy_photo[,,1]=0
noisy_photo[,,2]=0
noisy_photo[,,3]=noisy_photo_blue
noisy_photo=ifelse(noisy_photo>1,1,noisy_photo)
plot(0:256,0:256,type="n",axes=TRUE)
rasterImage(noisy_photo,0,0,256,256)
```

```r
#We transform our noisy image to a greyscale image. We also print dimension of this greyscale image and print this image.
noisy_photo_grey=(noisy_photo_red+noisy_photo_green+noisy_photo_blue)/max(noisy_photo_red+noisy_photo_green+noisy_photo_blue)
plot(0:256,0:256,type="n",axes=TRUE)
dim(noisy_photo_grey)
rasterImage(noisy_photo_grey,0,0,256,256)
```

```r
#We produce our patches matrix with right number of instances and features. We also fill it with the values of patches.
patches=matrix(0,232*232,25*25)
k=1
for (i in 1:232){
    for (j in 1:232){
        patches[k,]=as.vector(noisy_photo_grey[(i):(i+24),(j):(j+24)])
        k=k+1
    }
}
```

```r
#We apply pca on our patches matrix. We see that variance covered by first component is remarkably bigger than the others.
pca=princomp(patches)
plot(pca)
```

```r
#We see that approximately 60% of variance is covered by the first component. The second component covers approximately 8%
#of the variance and the third one covers approximately 6% of variance.
summary(pca)
```

```r
#Let's normalise first score values and print them as an image. We also print the dimension.
normalised_scores_1=(pca$scores[,1] - min(pca$scores[,1]))/(max(pca$scores[,1]) - min(pca$scores[,1]))
dim(matrix(normalised_scores_1,232,232))
plot(0:232,0:232,type="n",axes=TRUE,main = "First Component")
rasterImage(t(matrix(normalised_scores_1,232,232)),0,0,232,232)
```

```r
#Let's normalise second score values and print them as an image. We also print the dimension.
normalised_scores_2=(pca$scores[,2] - min(pca$scores[,2]))/(max(pca$scores[,2]) - min(pca$scores[,2]))
dim(matrix(normalised_scores_2,232,232))
plot(0:232,0:232,type="n",axes=TRUE,main = "Second Component")
rasterImage(t(matrix(normalised_scores_2,232,232)),0,0,232,232)
```

```r
#Let's normalise third score values and print them as an image. We also print the dimension.
normalised_scores_3=(pca$scores[,3] - min(pca$scores[,3]))/(max(pca$scores[,3]) - min(pca$scores[,3]))
dim(matrix(normalised_scores_3,232,232))
plot(0:232,0:232,type="n",axes=TRUE,main = "Third Component")
rasterImage(t(matrix(normalised_scores_3,232,232)),0,0,232,232)
```

We see that the first component gives more details than the second and the third component.

```r
#Let's print the dimension for the first three eigen vectors and print the first rows of them.
dim(pca$loadings[,1:3])
head(pca$loadings[,1:3])
```

```r
#Let's normalise first eigen vectors and print it as an image. We also print the dimension.
normalised_loadings_1=(pca$loadings[,1] - min(pca$loadings[,1]))/(max(pca$loadings[,1]) - min(pca$loadings[,1]))
dim(matrix(normalised_loadings_1,25,25))
plot(0:25,0:25,type="n",axes=TRUE,main = "First Eigen Vector")
rasterImage(t(matrix(normalised_loadings_1,25,25)),0,0,25,25)
```

```r
#Let's normalise second eigen vector and print it as an image. We also print the dimension.
normalised_loadings_2=(pca$loadings[,2] - min(pca$loadings[,2]))/(max(pca$loadings[,2]) - min(pca$loadings[,2]))
dim(matrix(normalised_loadings_2,25,25))
plot(0:25,0:25,type="n",axes=TRUE,main = "Second Eigen Vector")
rasterImage(t(matrix(normalised_loadings_2,25,25)),0,0,25,25)
```

```r
#Let's normalise third eigen vector and print it as an image. We also print the dimension.
normalised_loadings_3=(pca$loadings[,3] - min(pca$loadings[,3]))/(max(pca$loadings[,3]) - min(pca$loadings[,3]))
dim(matrix(normalised_loadings_3,25,25))
plot(0:25,0:25,type="n",axes=TRUE,main = "Third Eigen Vector")
rasterImage(t(matrix(normalised_loadings_3,25,25)),0,0,25,25)
```

The pixels in the first eigen vector are more homogeneous and there is a cleaner view. However, the dark areas in the second and in the third eigenvectors are bigger and there is a more heterogeneous view. We can think that this situation causes a more uncertain and non clean view that we see in the second and in the third component.
