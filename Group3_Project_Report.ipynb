# B. Related Literature 

Karlis and Ntzoufras (1998) research on finding a statistical model which can predict the outcomes of soccer games in a classical way. They use Greek League data for season 1997-1998. The features which they use are classified into three classes, namely offensive parameters, defensive parameters and home effect. After comparing different statistical models, they find that Poisson distribution suffices for describing games. 

Suzuki et al. (2006) propose a Bayesian methodology for predicting match outcomes. The methodology is illustrated on the 2006 Soccer World Cup. The features they use in the research are classified into two groups, namely specialists’ opinions and the FIFA ratings. The method is applied to calculate the outcome probabilities at each match and also to simulate the whole competition in order to estimate classification probabilities in group stage and winning tournament chances for each team. Method is evaluated according to the DeFinetti measure and the percentage of correct forecasts.

Graham and Stott (2008) evaluate rationality of bookmaker (William Hill) odds with comparing the odds and ordered probit model, which they build to generate predictions for English football matches. Features which they use are team ratings, home advantage, distance travelled, bookmaker's odds. They show that there are systematic biases in bookmaker odds, and that these biases cannot be explained by William Hill odds omitting valuable, or excluding extraneous, information.

Langseth (2013) tries to find a statistical model which, based on the game-history so far in a season, can predict the outcome of next round's matches. After predicting, s/he considers the use of different money management techniques to evaluate and compare three models they use, namely static, dynamic and data-intensive. In addition, s/he also compares three different betting strategies. Therefore, s/he tries to maximize profit of bookmaker whatever the outcome.

Shi et al. (2013) explore whether models used are more significant than features using to build model or not. They evaluate factors of team success and adjusted efficiencies as features. They compare the models and features by looking performance of models. As a result, they show that choosing the right features gives the best results, so feature selection is more significant.

Louzada et al. (2014) try to estimate the average goals scored by each team by assuming that the number of goals scored by a team in a match follows a univariate Poisson distribution. They consider linear models which, with good predictive quality, easy implementation and low computational effort, express the sum and the difference of goals scored. The features they use are the goal average in a match, the home-team advantage, the team's offensive power, the opponent team's defensive power and a crisis indicator. The methodology is applied to the 2008-2009 English Premier League.

Bunker and Thabtah (2019) provide a critical analysis of the literature in ML, focusing on the application of Artificial
Neural Network (ANN) to sport results prediction. They propose a novel sport prediction framework through which ML can be used as a learning strategy. Some features they use in the research are results of historical matches, player performance indicators and opposition information. 

# C. Approach

In this project, we have five data sets, namely bets, booking, goals, matches and statistics. Bets data includes bookmaker with odd type, odd value and odd epoch for a match. Booking data includes fault, card and time information of home and away teams for each match. Goals data includes scorer and time information of each goal information for each match. Matches data includes match status, away and home teams, extra time, score, and league id information for each match. Statistics data includes match detailed statistics such as ball possession, corners, fouls, offsides, shots, cards for each matches' home and away team. As a result, data pre-processing is significant for this project since there are lots of different data sets.

### Interpreting the Methodology and Results

#### Some Methods
Multinomial logistic regression tends to find linear boundaries of classes, therefore when there is a non linearity, it might have lower performance. 

The same is valid for Lasso Regression. Even with elimination of irrelevant features we overcome some of the curse of dimentionality but the classification is stuck at a limit.  

On the other hand, SVM also tries to seperate the classes. However, with data at hand which can be considered as limited (because of number of matches and data at hand are limited) and which has multi-class responses, this approach also might not grasp the underlying classifier in detail.  

Stochastic gradient boosting trees have low RPS when large number of trees and more depth is used. However as the depth increases, RPS values become close to RPS of RF. However the performance of the algorithm decreases, also we face overfitting. RF is better in terms of performance and RPS. Therefore, stochastic gradient boosting trees method is not selected.  

#### What is the advantage of RF?
In our feature set, none explains most of the variance. Therefore some features might also present some noise. In general, logistic regression performs better when the number of noise variables is less than or equal to the number of explanatory variables and random forest has a higher true and false positive rate as the number of explanatory variables increases in a dataset. Logistic regression and random forest are comparable for smaller datasets with less than 1000 observations.[8]
Therefore, RF or tree based approaches have advantage over logistic regression.

#### Why does RF and Average of RF and RF Extended has best overall predictions? 
Random forest has best fit in overall models also when combined with extended random forest, it has the best results. For most of the predictions in several rounds, Average of RF and RF extended have the best predictions. On the contrary, in some rounds random forest predicts a little bit better than the average.

#### Why do we use Average of RF and RF Extended, instead of RF?
In the case of RF, tie probabilities are especially small when away team is stronger than the home team, and there are none or very few ties in previous games. However, because of the ordinality nature of the classes, if hometeam win and away team win probabilities are both high, and tie probability should not be too small. Because for away team or hometeam to make a comeback, they have to be in tie state in between. However, RF considers the classes as nominal. This makes interpretability of RF hard. On the other hand RPS values of RF are very low. So this can be considered as overfitting.

In particular, trees that are grown very deep tend to learn highly irregular patterns: they overfit their training sets, i.e. have low bias, but very high variance. Random forests are a way of averaging multiple deep decision trees, trained on different parts of the same training set, with the goal of reducing the variance. [9]

When RF extended is used, RPS values are closer to RF, and the predictions are more interpretable because of tie probabilities. So to eliminate very low tie percentages, average of the RF and RF extended is used. This has best RPS values and high interpretability.

Tree based approaches have lower RPS than other methods in general.
